```python
import cv2
import numpy as np
import winsound
# from pygame import mixer
from keras.models import load_model
```

```python
# mixer.init()
# sound = mixer.Sound('alarm.wav')
```

```python
model=load_model("densnet.h5")
sound_frequency=3000  # in hertz
sound_duration = 4000 # 3 seconds
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

video = cv2.VideoCapture(0)
counter = 0

while True:
    flag,image = video.read()
    if (flag==False):
        break
     
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray,scaleFactor=1.2,minNeighbors=3,)
    eye_cascade = cv2.CascadeClassifier("haarcascade_eye_tree_eyeglasses.xml")
    

    for x,y,w,h in faces:

        roi_gray = gray[y:y+h,x:x+w]
        roi_color = image[y:y+h,x:x+w]
#         cv2.imwrite(f"{counter}.jpg",roi_color)
#         counter +=1
        cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,255),2)

        eyes = eye_cascade.detectMultiScale(roi_gray,scaleFactor=1.2,minNeighbors=3)

        if len(eyes)==0:
            print("eyes not detected")
        else:  
            print(" detected")
            for (ex,ey,ew,eh) in eyes:
                cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,255),2)
                eyes_roi = roi_color[ey:ey+eh,ex:ex+ew]

    if len(faces)==0:
            print("Face not detected")
            cv2.putText(image,"Face Not Detected",(200,70),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),2,cv2.LINE_4)
            
    else:     
        input_eye=cv2.resize(eyes_roi,(80,80))
#         print(len(input_eye))
        input_eye = np.expand_dims(input_eye,axis=0)
        input_eye=input_eye/255
        prediction=model.predict(input_eye)
        print(prediction)

        if prediction[0][0]>=0.5:
            status = "Opened Eyes"
            counter = 0
            cv2.putText(image,status,(200,70),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2,cv2.LINE_4)

            # Black Background rectangle
            cv2.rectangle(image,(0,0),(170,70),(0,0,0),-1)
            cv2.putText(image,f"Count :{counter} ",(10,30),cv2.FONT_HERSHEY_COMPLEX,0.6,(255,255,255),2,cv2.LINE_4)

        else:
            counter+=1
            status = "Closed"
            cv2.putText(image,status,(200,70),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2,cv2.LINE_4)
            
            cv2.rectangle(image,(0,0),(170,70),(0,0,0),-1)
            cv2.putText(image,f"Count :{counter} ",(10,30),cv2.FONT_HERSHEY_COMPLEX,0.6,(0,0,255),2,cv2.LINE_4)

            if counter>=20:
                winsound.Beep(frequency=sound_frequency,duration=sound_duration)
                counter = 0

    

    cv2.imshow("Drowsiness Detection System",image)
    key=cv2.waitKey(25)
    if(key==ord("q")):
        break
        
video.release()
cv2.destroyAllWindows()

```
